```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
%config InlinBackend.figure_format="retina"
plt.rcParams['font.sans-serif']=['SimHei'] #
plt.rcParams['axes.unicode_minus']=False #

# 
def histogram(image):
    (row, col) = image.shape
    #256list
    hist = [0]*256
    for i in range(row):
        for j in range(col):
            hist[image[i,j]] += 1
    return hist

image0 = cv2.imread(r'..\img\train6.jpg',0)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.axis("off")
plt.imshow(image0,vmin=0, vmax=255,cmap = plt.cm.gray)
# plt.title('')
image_hist0 = histogram(image0)
plt.subplot(1,3,2)
plt.xlabel("")
plt.ylabel("")
plt.plot(image_hist0)
plt.subplot(1,3,3)
# plt.title('')
plt.bar(range(256),image_hist0)
plt.xlabel("")
plt.ylabel("")
plt.show()
```
