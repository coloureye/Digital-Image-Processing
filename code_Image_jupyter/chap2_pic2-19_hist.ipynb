```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
%config InlinBackend.figure_format="retina"
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负

# 统计各灰度值的像素个数
def histogram(image):
    (row, col) = image.shape
    #创建长度为256的list
    hist = [0]*256
    for i in range(row):
        for j in range(col):
            hist[image[i,j]] += 1
    return hist

image0 = cv2.imread(r'..\img\train6.jpg',0)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.axis("off")
plt.imshow(image0,vmin=0, vmax=255,cmap = plt.cm.gray)
# plt.title('原图像')
image_hist0 = histogram(image0)
plt.subplot(1,3,2)
plt.xlabel("灰度值")
plt.ylabel("像素个数")
plt.plot(image_hist0)
plt.subplot(1,3,3)
# plt.title('柱形直方图')
plt.bar(range(256),image_hist0)
plt.xlabel("灰度值")
plt.ylabel("像素个数")
plt.show()
```
