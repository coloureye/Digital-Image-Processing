```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
%config InlinBackend.figure_format="retina"
plt.rcParams['font.sans-serif']=['SimHei'] #
plt.rcParams['axes.unicode_minus']=False #

img = cv2.imread(r'..\img\alphabet.jpg',0)
# 
dx=dy=50
rows,cols= img.shape[:2]
M1 = np.float32([[1,0,dx],[0,1,dy]])  #
dst1 = cv2.warpAffine(img,M1,(cols,rows))
# 
M2=cv2.getRotationMatrix2D((cols/2,rows/2),30 ,1)
dst2=cv2.warpAffine(img,M2,(cols,rows))
# 
M3 = np.float32([[1,0,0],[0,-1,rows]])
dst3 = cv2.warpAffine(img,M3,(cols,rows))
plt.figure(figsize=(12,5))
plt.subplot(141)
plt.imshow(img,cmap="gray")
plt.axis("off")
plt.subplot(142)
plt.imshow(dst1,cmap="gray")
plt.axis("off")
plt.subplot(143)
plt.imshow(dst2,cmap="gray")
plt.axis("off")
plt.subplot(144)
plt.imshow(dst3,cmap="gray")
plt.axis("off")
plt.show()
```
