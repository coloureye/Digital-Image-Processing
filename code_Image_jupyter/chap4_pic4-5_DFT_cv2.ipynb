```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
%config InlinBackend.figure_format="retina"
plt.rcParams['font.sans-serif']=['SimHei'] #
plt.rcParams['axes.unicode_minus']=False #

#Numpy
img = cv2.imread(r'..\img\peppers.bmp',0)
dft = np.fft.fft2(img)
#dft[:,:,0],dft[:,:,1]
magnitude0= np.abs(dft)
magnitude1= 20*np.log(1+magnitude0)#
phase_angle0=np.angle(dft)
img_back=np.fft.ifft2(dft)
img_back1 = np.abs(img_back)

plt.figure(figsize=(10,6))
plt.subplot(151),plt.imshow(img, cmap = 'gray')
plt.title(''), plt.axis('off')
plt.subplot(152),plt.imshow(magnitude0, cmap = 'gray')
plt.title(''), plt.axis('off')
plt.subplot(153),plt.imshow(magnitude1, cmap = 'gray')
plt.title(''), plt.axis('off')
plt.subplot(154),plt.imshow(phase_angle0, cmap = 'gray')
plt.title(''), plt.axis('off')
plt.subplot(155),plt.imshow(img_back1, cmap = 'gray')
plt.title(''),plt.axis('off')
plt.savefig("ch4-5.jpg")
# plt.show()
```
