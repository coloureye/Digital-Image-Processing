```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
%matplotlib inline
%config InlinBackend.figure_format="retina"
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负
```

```python
img0 = cv2.imread(r'..\img\light_circle.jpg',0)
_,img_seg=cv2.threshold(img0,128,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
kernalSize=19
img_seg_adapt = cv2.adaptiveThreshold(img0, 255, cv2.ADAPTIVE_THRESH_MEAN_C, \
                                   cv2.THRESH_BINARY, kernalSize, 6)
plt.figure(figsize=(15,5))
plt.subplot(131)
plt.imshow(img0,cmap="gray")
plt.axis("off")
plt.subplot(132)
plt.imshow(img_seg,cmap="gray")
plt.axis("off")
plt.subplot(133)
plt.imshow(img_seg_adapt,cmap="gray")
plt.axis("off")
plt.show()
```

```python
#调参代码
kernalSize=7
plt.ion()
for i in range(10):
    c=4
    for j in range(10):
        plt.cla()
        img_seg_adapt = cv2.adaptiveThreshold(img0, 255, \
            cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, kernalSize, c)
        plt.imshow(img_seg_adapt,cmap="gray")
        print(kernalSize, c)
        plt.pause(0.01)
        c+=2
    kernalSize+=2
plt.show()
```
