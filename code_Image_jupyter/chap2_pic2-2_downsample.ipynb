```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

#图像下采样，其中scale是下采样比例
def downSample(img,scale):    
    num=int(1//scale)     #获得间隔像素值
    m,n=img.shape
    mm=int(m*scale)       #下采样后图像大小
    nn=int(n*scale)
    img1 = np.zeros((mm,nn), np.uint8)
    ii=0
    for i in range(0,m,num):   
        img1[ii,...]=img[i,range(0,n,num)]
        ii+=1
    return img1.copy()

img0 = cv2.imread('../img/rose.tif',0); 
plt.figure(figsize=(12,6))
plt.subplot(2,4,1)
plt.axis("off")
plt.imshow(img0,cmap="gray")
m,n=img0.shape
plt.title('(a) '+str(m)+'*'+str(n))

#下采样7次后的图像大小
for iter in range(2,9):
    img0=downSample(img0.copy(),0.5)
    plt.subplot(2,4,iter)
    plt.axis("off")
    plt.imshow(img0,cmap="gray")
    m,n=img0.shape
    plt.title('('+chr(ord('a')+iter-1)+') '+str(m)+'*'+str(n))
plt.show() 
```
