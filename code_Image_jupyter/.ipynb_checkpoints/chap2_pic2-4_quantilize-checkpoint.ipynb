```python
# -*- coding: utf-8 -*-
# 降低图像量化等级
import cv2  
import numpy as np  
import matplotlib.pyplot as plt
%matplotlib inline
#用来正常显示中文标签
plt.rcParams['font.sans-serif']=['SimHei']

def quantilize(img0,scale):
    img=img0.copy()
    #获取图像高度和宽度
    height = img.shape[0]
    width = img.shape[1]
    #创建一幅图像
    new_img = np.zeros((height, width), np.uint8)
    gap=256//scale
#     print("gap=",gap)
    for i in range(height):
        for j in range(width):
            num=0
            while num<scale:      
                if num*gap<=img[i, j]<(num+1)*gap :
                    gray =num*gap                    
                    break
                num+=1        
            new_img[i, j] = np.uint8(gray)
    return new_img

if __name__=="__main__":
    #读取原始图像
    img0 = cv2.imread('./img/skull.jpg',0)
    
    #设置量化级数（0-255之间的2次幂）
    scales=[128,64,32,16,8,4,2]
    plt.figure(figsize=(10,6))
    plt.subplot(2,4,1)
    plt.axis("off")
    plt.imshow(img0,cmap=plt.cm.gray)
    plt.title("原图")
    i=2
    for scale in scales:
        new_img=quantilize(img0,scale)
#         print(np.max(img0))
#         print(np.max(new_img))
        plt.subplot(2,4,i)
        plt.axis("off")
        plt.imshow(new_img,cmap=plt.cm.gray)
        plt.title("量化级="+str(scale))
        i+=1
    plt.show()
```
